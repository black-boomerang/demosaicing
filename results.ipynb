# Результаты восстановления изображения из CFA

Время работы алгоритма на изображении: **1476.562 секунды**

Время работы на одном мегапикселе: **171.059 секунды**

PSNR: **25.179**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
original_lum = cv2.imread('OriginalLuminance.bmp', cv2.IMREAD_GRAYSCALE)[410:960, 2300:2600]
recovered_lum = cv2.imread('RecoveredLuminance.bmp', cv2.IMREAD_GRAYSCALE)[410:960, 2300:2600]
```

```python
plt.figure(figsize=(18, 20))
plt.subplot(121)
plt.imshow(original_lum, cmap='gray')
plt.axis('off')
plt.title('Яркость оригинала')
plt.subplot(122)
plt.imshow(recovered_lum, cmap='gray')
plt.axis('off')
plt.title('Яркость восстановленного изображения')
plt.show()
```

По такой картинке сложно судить об изменении разрешения. Поэтому нарисуем для них графики контура.

```python
plt.figure(figsize=(18, 20))
plt.subplot(121)
plt.contour(original_lum[::-1], cmap='gray')
plt.axis('off')
plt.title('Яркость оригинала')
plt.subplot(122)
plt.contour(recovered_lum[::-1], cmap='gray')
plt.axis('off')
plt.title('Яркость восстановленного изображения')
plt.show()
```

По этим графикам видно, что линии для оригинала начинают сливаться на отметке около 13.5, а для восстановленного изображения в районе 10.5. То есть разрешение упало примерно в 1.22 раза. Проделаем то же самое для горизонтальных линий.

```python
original_lum = cv2.imread('OriginalLuminance.bmp', cv2.IMREAD_GRAYSCALE)[980:1210, 2200:2700]
recovered_lum = cv2.imread('RecoveredLuminance.bmp', cv2.IMREAD_GRAYSCALE)[980:1210, 2200:2700]
```

```python
plt.figure(figsize=(18, 12))
plt.subplot(211)
plt.contour(original_lum[::-1], cmap='gray')
plt.axis('off')
plt.title('Яркость оригинала')
plt.subplot(212)
plt.contour(recovered_lum[::-1], cmap='gray')
plt.axis('off')
plt.title('Яркость восстановленного изображения')
plt.show()
```

Для горизонтальных линий наблюдаем аналогичную картину.

Снижение разрешения восстановленного изображения по сравнению с оригиналом $\approx$ **1.22**
